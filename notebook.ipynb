```r
install.packages("BiocManager")

BiocManager::install("AnnotationHub")
BiocManager::install("ensembldb")
```

```r
library(AnnotationHub)
library(ensembldb)

ah <- AnnotationHub()
edb <- ah[["AH119325"]]

# Transcripts
tx <- transcripts(edb, filter = ~ gene_id == "ENSG00000122585")

write.table(tx,
            file = "SP7_transcripts.tsv",
            sep = "\t",
            row.names = FALSE,
            quote = FALSE)

# Proteins
pr <- proteins(edb, filter = ~ gene_id == "ENSG00000122585")

write.table(pr,
            file = "SP7_proteins.tsv",
            sep = "\t",
            row.names = FALSE,
            quote = FALSE)

# Mapping
tx_ids <- pr$tx_id
tx_gr <- transcripts(edb, filter = ~ tx_id %in% tx_ids)

mapping_df <- as.data.frame(tx_gr)

write.table(mapping_df,
            file = "SP7_protein_genome_map.tsv",
            sep = "\t",
            row.names = FALSE,
            quote = FALSE)
```
